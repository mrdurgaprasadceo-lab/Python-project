Let's say we have a list L1=[1,2,2,3,2,3,5] now identify the most reputation element in the list using dictinory in python 


```python
# 1s -1
# 2s -2
# 3s -3
# 4s -4
# 5s -5
```

```python
num_counter = {}
```

```python
# {1:1, 2:2, 3:2, .....}
```

```python
L1 = [1, 2, 2, 3, 2, 3, 4, 5, ]
for i in L1 :
    if i in num_counter:
        num_counter [i] +=1
    else:
        num_counter [i] =1
```

```python
num_counter
```

```python
temp = 0
final_result = None
for k, v in num_counter.items ():
    print(f"k: {k}, v: {v}, tem: {temp}")
    if v >temp:
        temp = v
        final_result = k
```
