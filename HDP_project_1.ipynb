```python
print("how many task do you have today?")
num = int(input())
tasks = []
completed = []
incomplete = []

for i in range(num):
    task = input(f"Enter task {i+1}: ")
    tasks.append(task)

print("\nNow tell me which task you finished (yes or no):")
for task in tasks:
  ans = input(f"Did you finish task '{task}'? (yes or no): ")
  if ans.lower() == "yes":
    completed.append(task)
  else:
    incomplete.append(task)

print("\nCompleted tasks:", completed)
print("Incomplete tasks:", incomplete)
```
